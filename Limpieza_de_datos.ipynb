Podemos observar que en la columna *"Item_Weight"* y *"Outlet_Size"* presentamos valores faltantes

```python
df.info()
```

Llenamos los valores faltantes de 'Item_Weight' con la función '.interpolate' para dar un balance entre el valor de la fila de arriba del faltante con el valor de la fila de abajo del faltante.

```python
df.loc[df['Item_Weight'].isna()]

```

```python
df['Item_Weight']=df['Item_Weight'].interpolate(method='linear')
df.head(25)
```

```python
df.info()
```

Podemos observar valores mal escritos 'reg'

```python
df.loc[df['Item_Fat_Content']=='reg']
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('reg','Regular')
df
```

Observamos que hay diferentes valores con LF en vez de 'Low Fat'

```python
df.loc[df['Item_Fat_Content']=='LF']
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('LF','Low Fat')
```

```python
df.loc[df['Item_Fat_Content']=='LF']
```

Podemos observar valores mal escrito 'low fat'

```python
df.loc[df['Item_Fat_Content']=='low fat']
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('low fat','Low Fat')
```

```python
df.loc[df['Item_Fat_Content']=='low fat']
```

Trabajamos sobre los valores faltantes en la variable 'Outlet_Size'

```python
df['Outlet_Size'].isna()
```

```python
df.loc[df['Outlet_Size'].isna()]
```

```python
df['Outlet_Size']=df['Outlet_Size'].fillna(method='ffill')
```
